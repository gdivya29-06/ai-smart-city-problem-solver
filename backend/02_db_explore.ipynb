```python
import sqlite3, uuid
from datetime import datetime

DB_PATH = 'complaints.db'

def init_db():
    conn = sqlite3.connect(DB_PATH)
    conn.execute('''CREATE TABLE IF NOT EXISTS complaints (
        id TEXT PRIMARY KEY, issue TEXT, department TEXT,
        status TEXT, location TEXT, description TEXT, created_at TEXT)''')
    conn.commit()
    conn.close()

def save_complaint(issue, department, location, description):
    cid = str(uuid.uuid4())
    conn = sqlite3.connect(DB_PATH)
    conn.execute('INSERT INTO complaints VALUES (?,?,?,?,?,?,?)',
        (cid, issue, department, 'Complaint Registered', location, description, datetime.utcnow().isoformat()))
    conn.commit()
    conn.close()
    return cid

def get_all_complaints():
    conn = sqlite3.connect(DB_PATH)
    conn.row_factory = sqlite3.Row
    rows = conn.execute('SELECT * FROM complaints ORDER BY created_at DESC').fetchall()
    conn.close()
    return [dict(r) for r in rows]

init_db()
print('Database ready!')
```

```python
save_complaint('pothole', 'Road Maintenance', 'MG Road, Block 5', 'Large pothole near bus stop')
save_complaint('garbage', 'Waste Management', 'Park Street', 'Overflowing dustbin')
save_complaint('streetlight', 'Electricity Department', 'Lake Road', 'Light not working')
print('3 complaints saved!')
```

```python
import pandas as pd

all_complaints = get_all_complaints()
df = pd.DataFrame(all_complaints)
df[['issue', 'department', 'status', 'location']]
```
