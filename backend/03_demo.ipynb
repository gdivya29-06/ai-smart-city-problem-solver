```python
import subprocess, time, sys

server = subprocess.Popen(
    [sys.executable, '-m', 'uvicorn', 'main:app', '--port', '8000'],
    cwd=r'C:\Users\Manya\ai-smart-city-problem-solver\backend'
)

time.sleep(3)
print('Server started: http://127.0.0.1:8000')
print('Docs:           http://127.0.0.1:8000/docs')
```

```python
import requests
r = requests.get('http://127.0.0.1:8000/health')
print(r.json())
```

```python
with open('pothole.jpg', 'rb') as img:
    response = requests.post(
        'http://127.0.0.1:8000/report',
        data={'location': 'MG Road, Block 3'},
        files={'image': ('pothole.jpg', img, 'image/jpeg')}
    )
result = response.json()
for k, v in result.items():
    print(f'  {k}: {v}')
```

```python
import pandas as pd
r = requests.get('http://127.0.0.1:8000/complaints')
pd.DataFrame(r.json())[['issue', 'department', 'status', 'location']]
```

```python
server.terminate()
print('Server stopped.')
```
